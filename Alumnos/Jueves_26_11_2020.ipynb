```python
player_team.set_index('Player', inplace=True)
```

```python
def get_teamate(player_teams):
    lis_temates=[]
    for i in player_teams['Teams']['Player']:
        lis_temates= lis_temates+i
    list_ret = list( set(lis_temates) )
    #list_ret.remove(player_teams.name)
    return list_ret
```

```python
team_temates_se=player_team.apply(lambda y : get_teamate(y) , axis=1)
```

```python
team_temates_se
```

```python
team_temates_se.sort_index(inplace=True)
```

```python
for i in team_temates_se.index:
    #print(i)
    if type(i)!= float:
        team_temates_se[i].remove(i)
```

```python
lis_to=[]
for i in team_temates_se.index:
    if type(i)!= float:
        dic_pa={'Player':i}
        for j in team_temates_se[i]:
            dic_pa[j]=1
    lis_to.append(dic_pa)
```
